````python
from selenium import webdriver
from bs4 import BeautifulSoup
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
wd = webdriver.Chrome('chromedriver',options=options)
wd.get("https://news.naver.com")
pageHtml = wd.page_source
NN = BeautifulSoup(pageHtml, "html.parser")

contents = NN.find_all('div', {'class':'hdline_article_tit'})

for value in contents:
  result = value.find('a')
  title = result.text
  link = result.get('href')
  print(title.strip())
  print(link.strip())
````

## - 빌보드 차트 1위에서 100위까지의 순위를 크롤링하여 나열하시오.

site : ``` http://billboard.co.kr/chart/week/ ```

결과예시  
```
라일락  |  아이유(IU)
롤린 (Rollin')  |  브레이브걸스(Brave Girls)

Dynamite  |  방탄소년단

Timeless  |  SG워너비
On The Ground  |  로제(ROSÉ)
봄 안녕 봄  |  아이유(IU)
```

````python
#@title
from selenium import webdriver
from bs4 import BeautifulSoup
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
wd = webdriver.Chrome('chromedriver',options=options)
wd.get("http://billboard.co.kr/chart/week/")
pageHtml = wd.page_source
BC = BeautifulSoup(pageHtml, "html.parser")

print(BC)
````

````python
#@title
contents = BC.find_all('div', {'class':'chart_tit_tt'})

# print(contents)
for value in contents:
  title = value.find('p', {'class':'chart_tit'}).text
  singer = value.find('span', {'class':'chart_artist'}).text
  
  print(title.strip(), ' | ', singer.strip())
````

````python
#@title
from selenium import webdriver
from bs4 import BeautifulSoup
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
wd = webdriver.Chrome('chromedriver',options=options)
wd.get("https://dhlottery.co.kr/gameResult.do?method=byWin")
pageHtml = wd.page_source
LN = BeautifulSoup(pageHtml, "html.parser")

contents = LN.find_all('div', {'class':'win_result'})

print(contents)

#당첨회차

for value in contents:
  round = value.find('h4').text
  num1 = value.find('span', {'class':'ball_645 lrg ball1'}).text
  num2 = value.find('span', {'class':'ball_645 lrg ball2'}).text
  num3 = value.find('span', {'class':'ball_645 lrg ball3'}).text
  num4 = value.find('span', {'class':'ball_645 lrg ball4'}).text
  num5 = value.find('span', {'class':'ball_645 lrg ball5'}).text
  num6 = value.find('div', {'class':'num bonus'}).find('span', {'class':'ball_645 lrg ball2'}).text

  print(round)
  print("당청번호: ", num1, num2, num3, num4, num5)
  print("보너스 번호: ", num6)
````
